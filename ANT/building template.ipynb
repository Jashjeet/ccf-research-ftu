```
import ants
import os

import warnings
warnings.filterwarnings(action='ignore')

data_dir='./'
```

##### Loading the data

```
files=[i for i in os.listdir(os.getcwd()) if '.nii.gz' in i]

target_img=ants.image_read('/content/VHF_Kidney_L_High.nii.gz')
population=[ants.image_read(i).resample_image_to_target(target_img) for i in files]
```

##### Building the template

```
btp = ants.build_template( image_list = population,
                          synMetric = 'SyN')
```

```
btp
```

##### Displaying the constructed template

```
btp.plot()
```

```
ants.image_write(btp,'built_template.nii.gz')
```
