# <center> Rigid Registration</center>

## 0% (No) Noise

```
fixed=ants.image_read(os.path.join(data_dir,'0_percent_fixed.nii.gz'))
moving=ants.image_read(os.path.join(data_dir,'0_percent_moving.nii.gz'))
```

##### Displaying the moving image

```
moving.plot()
```

##### Registering the moving image in the fixed image

```
mytx = ants.registration(fixed=fixed, moving=moving, type_of_transform = 'DenseRigid')
mytx
```

### Displaying the results

##### Registration Results

```
fixed.plot(overlay=mytx['warpedmovout'],title='Fixed image overlaid with registered moving image')
fixed.plot(overlay=mytx['warpedmovout'],title='Registered moving image',black_bg=False)
```

##### Warped Image

```
warped = ants.apply_transforms( fixed=fixed, moving=moving, transformlist=mytx['fwdtransforms'])
warped.plot(title='Warped Moving Image')
```

```
ants.image_write(warped,os.path.join(data_dir,'rigid_warped_0_percent.nii.gz'))
```

# <center>Comparing Registration results</center>

## 0% (No) Noise

```
fixed=ants.image_read(os.path.join(data_dir,'0_percent_fixed.nii.gz'))
moving=ants.image_read(os.path.join(data_dir,'0_percent_moving.nii.gz'))

rig=ants.registration(fixed=fixed, moving=moving, type_of_transform = 'DenseRigid')
syn=ants.registration(fixed=fixed, moving=moving, type_of_transform = 'SyNAggro')

fixed.plot(overlay=rig['warpedmovout'],title='Rigid Registration',black_bg=False)
fixed.plot(overlay=syn['warpedmovout'],title='SyN Registration',black_bg=False)
```

```
warped = ants.apply_transforms(fixed=fixed, moving=moving, transformlist=rig['fwdtransforms'])
warped.plot(title='Rigid Registration: Warped Moving Image')

warped = ants.apply_transforms(fixed=fixed, moving=moving, transformlist=syn['fwdtransforms'])
warped.plot(title='SyN Registration: Warped Moving Image')
```
